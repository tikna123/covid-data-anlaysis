```python
import pandas as pd
df = pd.read_csv("covid_data.csv",parse_dates=['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['new_cases_by_population'] = df["new_cases"]/df["population"]
df['new_deaths_by_population'] = df["new_deaths"]/df["population"]
df.head()
df[df["new_cases"]<0].head()
```

# Country wise analysis

```python
df_location_wise = df.groupby('location')
```

```python
# no of new_cases by location..sorted
df_location_wise.agg({'new_cases':sum}).sort_values(by=['new_cases'], ascending=False)
# top 3 cases by day for each location
#df_location_wise['new_cases'].nlargest(3)
temp = df_location_wise.agg({"new_cases":['max','min'], "new_deaths":['max','mean','median']})
temp.head()
#temp['new_cases']['max']
```
